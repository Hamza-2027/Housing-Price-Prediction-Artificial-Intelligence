```python
# import necessary libraries
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

# load input dataset
data = pd.read_csv("/content/drive/MyDrive/Peshawar/new whole.csv")

# split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(data[['demand', 'grid1-load','grid1-loss', 'grid1-loss-prophet-daily','grid1-loss-prophet-pred','grid1-loss-prophet-trend','grid1-loss-prophet-weekly','grid1-loss-prophet-yearly','grid1-temp','season_x','season_y','month_x','month_y','week_x','week_y','weekday_x','weekday_y','holiday','hour_x']], data['hour_y'], test_size=0.3)

# train the model
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)

# predict on training and validation sets
y_train_pred = regressor.predict(X_train)
y_val_pred = regressor.predict(X_val)

# calculate RMSE, MSE, AMSE for training set
rmse_train = (mean_squared_error(y_train, y_train_pred))**0.5
mse_train = mean_squared_error(y_train, y_train_pred)
amse_train = mean_absolute_error(y_train, y_train_pred)

# calculate RMSE, MSE, AMSE for validation set
rmse_val = (mean_squared_error(y_val, y_val_pred))**0.5
mse_val = mean_squared_error(y_val, y_val_pred)
amse_val = mean_absolute_error(y_val, y_val_pred)


# print the results
print("Training set:")
print("RMSE: ", rmse_train)
print("MSE: ", mse_train)
print("AMSE: ", amse_train)


```

```python
print("Validation set:")
print("RMSE: ", rmse_val)
print("MSE: ", mse_val)
print("AMSE: ", amse_val)
```

```python
# Plot the training data
plt.scatter(y_train, y_train_pred, label='Training Data')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.title('Training Data')
plt.show()
```

```python
# Plot the validation data
plt.scatter(y_val, y_val_pred, label='Validation Data')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=2)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.title('Validation Data')
```

```python
# plot training and validation sets
plt.scatter(y_train, y_train_pred, color='blue', label='training set')
plt.scatter(y_val, y_val_pred, color='orange', label='validation set')
plt.xlabel("True values")
plt.ylabel("Predictions")
plt.legend()
plt.show()
```
