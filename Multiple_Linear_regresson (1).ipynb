```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# Input dataset
data = pd.read_csv('/content/drive/MyDrive/Peshawar data/Peshawar Circle Small drop.csv')

# Split dataset into training and validation sets
train_data, val_data = train_test_split(data, test_size=0.3)

# Define X and y for training data
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

# Define X and y for validation data
X_val = val_data.iloc[:, :-1]
y_val = val_data.iloc[:, -1]

# Create and fit the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict using the model
y_train_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)

# Calculate the RMSE, MSE, and AMSE for the training data
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
train_mse = mean_squared_error(y_train, y_train_pred)
train_amse = mean_absolute_error(y_train, y_train_pred)

# Calculate the RMSE, MSE, and AMSE for the validation data
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
val_mse = mean_squared_error(y_val, y_val_pred)
val_amse = mean_absolute_error(y_val, y_val_pred)

# Print the results
print("Training RMSE: ", train_rmse)
print("Training MSE: ", train_mse)
print("Training AMSE: ", train_amse)

```

```python
print("Validation RMSE: ", val_rmse)
print("Validation MSE: ", val_mse)
print("Validation AMSE: ", val_amse)
```

```python
# Plot the training data
plt.scatter(y_train, y_train_pred, label='Training Data')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.title('Training Data')
plt.show()
```

```python
# Plot the validation data
plt.scatter(y_val, y_val_pred, label='Validation Data')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'k--', lw=2)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.title('Validation Data')
```

```python
# Plot the training and validation data
plt.scatter(y_train, y_train_pred, label='Training Data')
plt.scatter(y_val, y_val_pred, label='Validation Data')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.legend()
plt.show()
```
